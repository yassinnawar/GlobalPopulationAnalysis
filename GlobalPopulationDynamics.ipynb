## Code Efficiency: Creating a Function for Plotting Population Growth

### In line with best programming practices, we aim for code that is clean, clear, and efficient. A key strategy to achieve this is the "Don't Repeat Yourself" (DRY) principle, which reduces redundancy and enhances readability and maintainability.

In keeping with the DRY principle, we've developed the function plot_population_growth to plot population growth for various countries across different time periods. Without this function, we would have to duplicate several lines of code each time we wanted to create a similar plot, but with a different set of countries or a different time period.

> By encapsulating this code in a function, we can easily generate as many of these plots as we need by simply calling the function with different arguments. This approach keeps our code base concise, easy to understand, and easy to manage.

>> In addition, encapsulating this code in a function also enhances code readability and makes it easier to identify and debug any issues that may arise. It simplifies our code structure, and makes it more straightforward to follow the flow of our data processing and analysis steps.

> We've also added a function to clean our data frame 

## Plotting Function Code⬇️

```python
import matplotlib.pyplot as plt

def plot_population_growth(countries, start_year, end_year, num_countries):
    # Create an empty plot
    fig, ax = plt.subplots(figsize=(12, 8))

    # Plot population growth for each country
    for country in countries[:num_countries]:
        country_data = populationData[populationData["Country Name"] == country]
        years = range(start_year, end_year + 1)  # Update this range with the desired start and end years
        populations = country_data.loc[:, str(years[0]):str(years[-1])].values[0]
        ax.plot(years, populations, label=country)

    # Add legend, title, and labels
    ax.legend()
    ax.set_title('Population Growth Over Time ({}-{})'.format(start_year, end_year))
    ax.set_xlabel('Year')
    ax.set_ylabel('Population')

    # Show the plot
    plt.tight_layout()
    plt.show()
```

## Data Cleaning Function Code⬇️

```python
def clean_data(df):
    filters = ['world', 'income', 'demographic', 'eastern', 'western', 'Asia', 
               'Euro', r'\bAfrica\b', r'\bAmerica\b', 'IBRD', 'members', 
               r'\bIDA\b', r'\bUN\b', 'countries', 'conflict','island','other','isle']

    for word in filters:
        df = df[~df['Country Name'].str.contains(word, case=False, na=False)]
                
    return df
```

### 5.1 ) Comparative Analysis of Relative and Absolute Population Growth

>#### To gain more nuanced insights into global population dynamics, we are extending our investigation into two additional aspects: relative population growth rate and absolute population increase. Each of these metrics offers unique perspectives into demographic trends and the factors driving them.





### Population Growth Calculation
Let's start by computing two different metrics for each country in our dataset: the total population growth and the growth rate from 1960 to 2022.

### The Population Growth is computed as the difference between the population in 2022 and 1960, representing the total increase in population over this period.

### The Growth Rate is calculated as the ratio of the Population Growth to the initial population in 1960, multiplied by 100 to obtain a percentage. This represents the relative increase in population size over the same period.

```python
# Calculate the absolute population increase
populationData['Population Growth'] = populationData['2022'] - populationData['1960']

# Calculate the growth rate
populationData['Growth Rate'] = (populationData['Population Growth'] / populationData['1960']) * 100
```

Now, let's use our computed metrics to identify the top 10 countries with the highest growth rate and the highest absolute population increase.

```python
# Sort the data by the growth rate in descending order
top_growth_countries = populationData.sort_values('Growth Rate', ascending=False).head(10)

# Sort the data by the absolute population increase in descending order
top_increase_countries = populationData.sort_values('Population Growth', ascending=False).head(10)
```

Next, we display these top 10 countries for both categories:

```python
# Print the top 10 countries with the highest growth rate and their population growth
print("Top 10 Countries with Highest Growth Rate:")
print(top_growth_countries[['Country Name', 'Growth Rate']])
print()
print("Top 10 Countries with Highest Absolute Population Increase:")
print(top_increase_countries[['Country Name', 'Population Growth']])
```

## <center> 5.5 ) Age Dependency Ratio Analysis <center>

### In this section, we will analyze the age dependency ratio, which provides insights into the balance between the economically productive population (typically aged between 15 to 64 years) and the dependent population (those under 15 and over 64).

### Age dependency ratio is a key indicator of the socioeconomic dynamics of a country and can influence factors such as social security systems, public health planning, and national economic models.

We will follow these steps:

    a. Calculate the dependency ratio for each country.
    b.Plot the dependency ratios to visualize the differences among countries.
    c.Discuss insights and potential implications.
    
>To calculate the dependency ratio, we'll use the following formula:

>> #### Dependency Ratio = ((Population aged 0-14 + Population aged 65+) / Population aged 15-64) * 100

>This ratio tells us the number of dependents per 100 working-age persons.


```python
# Calculate dependency ratios
pop_14['2022'] = pop_14['2022'].astype(float)
pop_15_64['2022'] = pop_15_64['2022'].astype(float)
pop_65['2022'] = pop_65['2022'].astype(float)

dependency_ratio = ((pop_14['2022'] + pop_65['2022']) / pop_15_64['2022']) * 100

# Add to the combined dataframe
combined['Dependency Ratio'] = dependency_ratio

# Sort the countries by dependency ratio in descending order
sorted_dependency_ratio = combined.sort_values('Dependency Ratio', ascending=False)

# Print top countries with highest dependency ratio
print(sorted_dependency_ratio.head())
```

## Handling Missing Data

#### As part of the data cleaning process, it's crucial to handle missing values in the dataset. In our GDP data, there are certain years for which GDP information is missing for some countries. These missing values can cause issues when performing subsequent data analysis or building models.

#### Therefore, we adopted the following approach to handle the missing GDP data:

1. **Linear Interpolation**: The idea behind interpolation is to use the values of the years preceding and following the missing year to estimate the value for that year. We used linear interpolation, which calculates the average of the last and next year's values to fill in the missing cell. This method assumes that the change in GDP from year to year is linear, which is a reasonable assumption for short periods.

2. **Median Imputation**: However, there might be situations where the missing year is the first or last year, meaning we don't have both a preceding and following year to use for interpolation. In these cases, we used median imputation, where the missing value is replaced with the median value of the available GDP values for that country. The median is a robust measure of central tendency, which means it's less affected by outliers in the data than the mean.

### By handling missing data, we ensure that our data analysis is based on a complete dataset and therefore more reliable.


```python
# Apply linear interpolation
for column in gdp_data.columns[1:]:  # We start from index 1 to skip the 'Country Name' column
    gdp_data[column] = gdp_data[column].interpolate(method='linear', limit_direction='both')

# Apply median imputation
for column in gdp_data.columns[1:]:
    gdp_data[column].fillna(gdp_data[column].median(), inplace=True)

print(gdp_data.head())
```
